```python
# line chart
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
sales2024 = [200,70,50,60,80]
sales2025 = [105,45,100,53,70]
sales2026 = [30,56,200,100,120]
#plt.figure(figsize=(10,8))
plt.plot(x, sales2024,label="Sales 2024", marker="o", color="blue")
plt.plot(x, sales2025,label="Sales 2025", marker="*", color="orange")
plt.plot(x, sales2026,label="Sales 2026", marker="x", color = "red")
plt.plot(x, [150,25,85,95,200],"g+--",label="Sales 2027")
plt.title("Sales through 2024-2026")
plt.xlabel("Sales Turnover")
plt.ylabel("Sales made")
plt.legend(loc = "upper right", bbox_to_anchor=(0.4,1.0))
plt.show()
```
